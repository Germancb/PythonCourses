```python
# % Importar librerias
import joblib
import pandas as pd
import numpy as np
```

```python
# 1. Cargar componentes
modelo = joblib.load('modelo_housing.pkl')
sc_X = joblib.load('escalador_X.pkl')
sc_y = joblib.load('escalador_y.pkl')
columnas_modelo = joblib.load('columnas_seleccionadas.pkl')
```

```python
# 2. ENTRADA DE USUARIO (Simulando que alguien escribe en una App)
# Aquí puedes poner "yes", "no" y el estado de la casa.
entrada_usuario = {
    'area': 5000,
    'bedrooms': 3,
    'bathrooms': 2,
    'stories': 2,
    'mainroad': 'yes',
    'guestroom': 'no',
    'hotwaterheating': 'no',
    'airconditioning': 'yes',
    'parking': 1,
    'prefarea': 'no',
    'furnishingstatus': 'semi-furnished' # Aunque RFE la descarte, la incluimos por lógica
}
```

```python
# 3. PREPROCESAMIENTO (Igual al entrenamiento)
# Convertir Yes/No a 1/0
for col in ['mainroad', 'guestroom', 'hotwaterheating', 'airconditioning', 'prefarea']:
    if col in entrada_usuario:
        entrada_usuario[col] = 1 if entrada_usuario[col].lower() == 'yes' else 0

# Convertir el diccionario a DataFrame
df_input = pd.DataFrame([entrada_usuario])
```

```python
# 4. SELECCIÓN DE COLUMNAS
# Filtramos solo las columnas que el modelo "aprendió" a usar (las 10 del RFE)
df_filtrado = df_input[columnas_modelo]
```

```python
# 5. ESCALADO
X_escalado = sc_X.transform(df_filtrado)
```

```python
# 6. CONVERTIR A DATAFRAME (Aquí estaba el error de nombre)
X_escalado_df = pd.DataFrame(X_escalado, columns=columnas_modelo)
```

```python
# 7. PREDICCIÓN Y DESESCALADO
# Cambiamos X_scaled_df por X_escalado_df
pred_decimal = modelo.predict(X_escalado_df) 

# Convertir de escala 0-1 a dinero real
precio_final = sc_y.inverse_transform(pred_decimal.reshape(-1, 1))

print(f"--- RESULTADO CON ENTRADA DE TEXTO ---")
print(f"PRECIO ESTIMADO: ${precio_final[0][0]:,.2f}")
```
