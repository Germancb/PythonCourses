```python
# 1 importar librerias
import pandas as pd
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
# 2. Carga directa
df = pd.read_csv("Housing.csv")
```

```python
# 3. Limpieza rápida (Esencial para que coincida)
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[varlist] = df[varlist].apply(lambda x: x.map({'yes': 1, "no": 0}))
status = pd.get_dummies(df['furnishingstatus'], drop_first=True)
df = pd.concat([df, status], axis=1)
df.drop(['furnishingstatus'], axis=1, inplace=True)
```

```python
# 4. Variables seleccionadas (El orden exacto que pediste)
columnas_rfe = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 
                'guestroom', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea']

X = df[columnas_rfe]
y = df[['price']]
```

```python
# 5. RE-AJUSTE TOTAL DE ESCALADORES
sc_X = MinMaxScaler()
sc_y = MinMaxScaler()

X_scaled = pd.DataFrame(sc_X.fit_transform(X), columns=columnas_rfe)
y_scaled = sc_y.fit_transform(y)
```

```python
# 6. ENTRENAMIENTO
modelo = LinearRegression()
modelo.fit(X_scaled, y_scaled)
```

```python
# 7. GUARDADO FORZADO (Sobrescribir archivos anteriores)
joblib.dump(modelo, 'modelo_housing.pkl')
joblib.dump(sc_X, 'escalador_X.pkl')
joblib.dump(sc_y, 'escalador_y.pkl')
joblib.dump(columnas_rfe, 'columnas_seleccionadas.pkl')

print("¡ARCHIVOS REGENERADOS! Ahora el escalador conoce el área máxima y mínima.")
```
